Set a pipeline for eaech model

```python
from transformers import pipeline


qa_nl_xlm = pipeline("question-answering", model="LouisJ444/Dutch_nlSQuAD", device=0, use_auth_token=auth_token)
qa_en_xlm = pipeline("question-answering", model="LouisJ444/xlm_roberta_en_SQuAD", device=0, use_auth_token=auth_token)
qa_nl_deberta = pipeline("question-answering", model="LouisJ444/dutchSquad_deberta", device=0, use_auth_token=auth_token)
qa_en_deberta = pipeline("question-answering", model="LouisJ444/squad_deberta", device=0, use_auth_token=auth_token)
qa_nl_bertje = pipeline("question-answering", model="LouisJ444/bertje_dutchsquad", device=0, use_auth_token=auth_token)
qa_en_distil = pipeline("question-answering", model="distilbert-base-cased-distilled-squad", device=0, use_auth_token=auth_token)
#squad_metric = load("squad")
```

Set the evaluator for QA

```python
from evaluate import evaluator
task_evaluator = evaluator("question-answering")
```

Go over each model and evaluate using the Test sets.

```python
# XLM English English
eval_results = task_evaluator.compute(
    model_or_pipeline=qa_en_xlm,
    data=squad["test"],
    metric="squad",
    squad_v2_format=False
)
print("XLM English > English SQuAD")
print(eval_results)

# XLM Dutch English
eval_results = task_evaluator.compute(
    model_or_pipeline=qa_nl_xlm,
    data=squad["test"],
    metric="squad",
    squad_v2_format=False
)
print("XLM Dutch > English SQuAD")
print(eval_results)

# Deberta English English
eval_results = task_evaluator.compute(
    model_or_pipeline=qa_en_deberta,
    data=squad["test"],
    metric="squad",
    squad_v2_format=False
)
print("DeBERTa English > English SQuAD")
print(eval_results)

# Deberta Dutch English
eval_results = task_evaluator.compute(
    model_or_pipeline=qa_nl_deberta,
    data=squad["test"],
    metric="squad",
    squad_v2_format=False
)
print("DeBERTa Dutch > English SQuAD")
print(eval_results)

# Distilbert English English
eval_results = task_evaluator.compute(
    model_or_pipeline=qa_en_distil,
    data=squad["test"],
    metric="squad",
    squad_v2_format=False
)
print("DistilBERT English > English SQuAD")
print(eval_results)

# Bertje Dutch English
eval_results = task_evaluator.compute(
    model_or_pipeline=qa_nl_bertje,
    data=squad["test"],
    metric="squad",
    squad_v2_format=False
)
print("Bertje Dutch > English SQuAD")
print(eval_results)

```
